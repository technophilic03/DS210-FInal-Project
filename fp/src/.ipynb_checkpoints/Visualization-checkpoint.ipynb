```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load graph data
graph_data = pd.read_csv('data/graph_data.csv')
edge_list = [(row["0"], row["1"]) for index, row in graph_data.iterrows()]

# Load centrality scores
centrality_data = pd.read_csv('data/centrality.csv')
centrality_dict = dict(zip(centrality_data['node_id'], centrality_data['score']))

# Load cluster data
cluster_data = pd.read_csv('data/clusters.csv')
cluster_dict = dict(zip(cluster_data['node_id'], cluster_data['cluster']))
```

```python
# Load graph data from CSV
df = pd.read_csv('data/graph_data.csv')
G = nx.from_pandas_edgelist(df, "0", "1")

# Use a simpler layout for large graphs
pos = nx.random_layout(G)  # This is much faster than spring_layout

# Draw the graph with simplified visual settings
nx.draw(G, pos, node_size=20, node_color='blue', with_labels=False)  # Turn off labels for faster rendering
plt.title('Network Graph - Simplified Layout')
plt.show()
```
